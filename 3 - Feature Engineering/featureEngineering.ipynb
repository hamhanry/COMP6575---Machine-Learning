```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import preprocessing
```

# Correlation

```python
dataset = pd.read_csv("/mnt/data/Project/Lecture/COMP6577-Machine-Learning/3 - Feature Engineering/conductivity.csv")
print(dataset.head())
print(dataset.corr())
```

# Heatmap Correlation

```python
import seaborn as sb
import matplotlib.pyplot as plt

plt.subplots(figsize=(10,5))
sb.heatmap(dataset.corr(), annot=True)
plt.show()
```

# Covariance

```python
print (dataset.cov())
```

# Heatmap Covariance

```python
plt.subplots(figsize=(10,5))
sb.heatmap(dataset.cov(), annot=True)
plt.show()
```

# Mean & Standard Deviation

```python
mean = dataset.mean()
std = dataset.std()
print("mean : ")
print(mean)
print("")
print("std")
print(std)
```

# Normalize Data

```python
X = dataset[["Temp", "Moisture", "Density"]]
y = dataset[["Conductivity"]]
# normalize the data attributes
normalized_X = preprocessing.normalize(X)

print("\n\n")
print(X)
print("")
print(normalized_X)
```

# Min Max Scaler

```python
scaler = preprocessing.MinMaxScaler()
scaler.fit(X)
print(scaler.transform(X))
```

```python
standardized_X = preprocessing.scale(X)

print("\n\n")
print(X)
print("")
print(standardized_X)
```
